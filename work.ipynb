## Top Pms by client satisfaction


```python
records.groupby(['department','region','priority']).agg({'client_satisfaction':'mean'}).reset_index()
```

```python
pm_rank=records.groupby('pm_name')['client_satisfaction'].mean().sort_values(ascending=False).reset_index()
```

```python
pm_rank['rank']=pm_rank['client_satisfaction'].rank(ascending=False, method='dense')
pm_rank[['pm_name', 'rank']].head()
```

## budget run rate per project type 


```python
projects['burn_rate']= projects['actual_spend_usd']/projects['budget_usd']
```

```python
burn_rate_by_type=(
    projects.groupby('project_type')['burn_rate'].mean().reset_index().sort_values('burn_rate', ascending=False)
)
burn_rate_by_type
```

## clients that  cancel the projects 


```python
cancelled_projects = projects[projects['status'] == 'Cancelled']
```

```python
cancel_counts=(
    cancelled_projects.groupby('client').size().reset_index(name='cancel_count').sort_values('cancel_count', ascending=False)
)
cancel_counts
```
